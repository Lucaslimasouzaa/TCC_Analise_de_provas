```python
import re
import os

# Lista de arquivos a serem processados (nomes das variáveis anteriormente usados para gerar arquivos .txt)
file_names = [
    "parsed_prova2015ve.txt", "parsed_prova2016in.txt", "parsed_prova2016ve.txt", "parsed_prova2017in.txt", "parsed_prova2017ve.txt",
    "parsed_prova2017pas1.txt", "parsed_prova2017pas2.txt", "parsed_prova2017pas3.txt", "parsed_prova2018ve.txt", "parsed_prova2018pas1.txt",
    "parsed_prova2018pas2.txt", "parsed_prova2018pas3.txt", "parsed_prova2019ve.txt", "parsed_prova2019pas1.txt", "parsed_prova2019pas2.txt",
    "parsed_prova2019pas3.txt", "parsed_prova2020ves.txt", "parsed_prova2022ves.txt", "parsed_prova2022pas1.txt", "parsed_prova2022pas2.txt",
    "parsed_prova2022pas3.txt", "parsed_prova2023in.txt", "parsed_prova2023ve.txt"
]

# Caminho para salvar o resultado consolidado
output_file_path = "/home/lucaslima/Dropbox/provas_vestibular/provas_txt/zipf_analysis.txt"

# Processa cada arquivo e salva os resultados
with open(output_file_path, "w") as output_file:
    for file_name in file_names:
        frequency = {}
        
        # Caminho completo do arquivo
        file_path = f"/home/lucaslima/Dropbox/provas_vestibular/provas_txt/{file_name}"
        
        # Verifica se o arquivo existe
        if os.path.exists(file_path):
            with open(file_path, "r") as content:
                text_string = content.read()
                
                # Extrai palavras e calcula frequência
                words = re.findall(r'\b[A-Za-z][a-z]{2,9}\b', text_string)
                for word in words:
                    count = frequency.get(word, 0)
                    frequency[word] = count + 1
                
                # Organiza por frequência em ordem decrescente
                most_frequent = dict(sorted(frequency.items(), key=lambda elem: elem[1], reverse=True))
                
                # Inicializa o contador para a frequência máxima
                top_count = 0
                
                # Grava o cabeçalho do arquivo
                output_file.write(f"Análise Zipf para {file_name}:\n")
                output_file.write("Palavra | Frequência | Relação Zipf\n")
                output_file.write("-" * 40 + "\n")
                
                # Itera sobre as palavras mais frequentes, limitando a 15
                for idx, (word, freq) in enumerate(most_frequent.items()):
                    if idx == 15:
                        break  # Limita a 15 palavras
                    if idx == 0:
                        top_count = freq
                    zipf_ratio = round(top_count / freq, 2)
                    output_file.write(f"{word} | {freq} | {zipf_ratio}\n")
                
                output_file.write("\n" + "=" * 40 + "\n\n")
                print(f"Análise de {file_name} concluída e salva.")
        else:
            print(f"Arquivo {file_name} não encontrado.")
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Caminho para o arquivo de texto
file_path = "/home/lucaslima/Dropbox/provas_vestibular/provas_txt/parsed_prova2015ve.txt"

# Inicializa o analisador de sentimento VADER
analyzer = SentimentIntensityAnalyzer()

# Lê o conteúdo do arquivo
with open(file_path, "r") as file:
    text = file.read()

# Realiza a análise de sentimento
score = analyzer.polarity_scores(text)

# Exibe o resultado
print(f"Sentimento geral:")
print(f"Positivo: {score['pos']:.2f}")
print(f"Neutro: {score['neu']:.2f}")
print(f"Negativo: {score['neg']:.2f}")
print(f"Score geral: {score['compound']:.2f}")
```
